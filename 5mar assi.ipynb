```python
'''
Q1. How can you create a Bokeh plot using Python code?
To create a Bokeh plot using Python code, you can use the bokeh.plotting module, which provides a high-level interface 
for creating a variety of plot types. '''
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

fig = figure()

fig.line(x, y)

# show the plot
output_notebook()
show(fig)
```
