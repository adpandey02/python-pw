```python
'''
Q1. Explain GET and POST methods.

GET and POST are HTTP methods used to send data from a client to a server.
GET method is used to request data from a server. The data is sent in the URL as a query string.
The request is visible in the URL, which can be bookmarked or shared easily. GET method is generally used for retrieving information.
POST method is used to submit data to a server. The data is sent in the request body and is not visible in the URL. 
POST method is generally used for submitting information.

----------------------------------------------------------------------------

Q2. Why is request used in Flask?

Flask is a micro web framework for building web applications in Python.
The request module in Flask is used to handle incoming requests from the client.
It provides an easy way to access the data sent in the request, such as form data, request headers, cookies, etc.
It also provides methods to check the type of request, like GET or POST, and to handle file uploads.

----------------------------------------------------------------------------

Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect the client to a different URL.
It is commonly used after a form submission or after a successful login to redirect the client to the appropriate page.
It helps to avoid duplicate form submissions by redirecting the client to a different URL after the form is submitted.

------------------------------------------------------------------------------

Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are HTML files that contain placeholders for dynamic content.
The render_template() function in Flask is used to render these templates and replace the placeholders with actual data.
It allows us to separate the presentation logic from the application logic, making the code more organized and easier to maintain.
The render_template() function takes the name of the template file and any arguments needed to fill in the placeholders.

-----------------------------------------------------------------------------
```
