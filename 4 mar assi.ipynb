```python
'''
Seaborn is a Python library based on Matplotlib that provides a higher-level interface for creating statistical visualizations. 
It offers many built-in plotting functions for various types of data and is especially useful for exploratory data analysis.

Here are five types of plots that can be created using Seaborn and their uses:

Scatter Plot - A scatter plot displays the relationship between two continuous variables using dots on a two-dimensional plane. 
It is used to visualize the correlation or association between variables and can be used to identify patterns or outliers.

Line Plot - A line plot is used to display trends in data over time or other ordered categories.
It can be used to show changes in a continuous variable over time, such as stock prices or temperature readings.

Bar Plot - A bar plot is used to compare categorical variables by displaying the values of one variable on the x-axis and the
count or aggregate of another variable on the y-axis. It can be used to show the distribution of data or compare the frequency
of different categories.

Box Plot - A box plot displays the distribution of a dataset using quartiles and outliers. 
It is used to identify the central tendency, spread, and skewness of a dataset, as well as to compare the distributions of multiple datasets.

Heatmap - A heatmap displays the values of a matrix using color intensity to represent the magnitude of the values.
It is used to identify patterns and relationships between variables, such as correlation or clustering,
and can be used to visualize complex data in a simple and intuitive way.
'''
```

# 2

```python
import seaborn as sns
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.lineplot(x=fmri.timepoint, y=fmri.signal, hue=fmri.event, style=fmri.region)
```

```python
sns.lineplot(x=fmri.timepoint, y=fmri.signal, hue=fmri.event, style=fmri.region, errorbar=None)
```
