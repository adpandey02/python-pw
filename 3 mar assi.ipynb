# 2

```python
import numpy as np
np.random.seed(3)
x = 3+  np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
'''
A scatter plot is a type of plot in which the values of two variables are plotted as points in a two-dimensional coordinate system, 
with one variable on the x-axis and the other variable on the y-axis. Each point in the plot represents an observation in the data set. 
Scatter plots are commonly used to visualize the relationship between two variables and to identify patterns or trends in the data.
In Matplotlib's Pyplot module, you can create a scatter plot using the scatter() function.
'''
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,3))
plt.scatter(x,y,c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('scatter plot')
plt.show()
```

```python

# Define the data for each line plot
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
fig, axs = plt.subplots(2, 2)

# Plot the data on each subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a main title to the figure
fig.suptitle('Four Line Plots')

# Display the figure
plt.show()
```

```python
'''
A box plot, also known as a box-and-whisker plot, is a type of plot that displays the distribution of a dataset using five summary 
statistics: the minimum value, the first quartile (Q1), the median, the third quartile (Q3), and the maximum value. 
The box represents the interquartile range (IQR), which is the range between the first and third quartiles, 
while the whiskers extend from the box to show the range of the data outside the IQR. Outliers may also be displayed as
individual points beyond the whiskers.'''
```

```python

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
labels = ['Box 1', 'Box 2']

plt.boxplot(data, labels=labels)
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Value')
plt.show()
```
